```python
for col in df.columns:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.show()
```

```python
# Plotting a heatmap to check for correlation between data
# Also the data is already encoded 
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

```python
churnplot = sns.pairplot(
            df,
            vars=['tenure', 'age', 'address', 'income'],
            hue='churn',
            plot_kws={'alpha':0.5}
            )
```
